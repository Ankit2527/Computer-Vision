```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Read the image
img = cv2.imread(r"D:\Dataset\ETHZ_PASCAL_Stickmen_v1.11\03\60903_1973-03-26_2009.jpg",1)
color = ('b','g','r')

for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)

plt.show()
```
