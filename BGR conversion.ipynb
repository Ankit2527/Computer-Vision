```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np

#Read the image
image = img.imread(r"D:\Dataset\ETHZ_PASCAL_Stickmen_v1.11\03\60903_1973-03-26_2009.jpg")
#Display the input image
plt.imshow(image)
```

```python
#Separating out red component
red_component = image[:,:,0]  
#Separating out green component
green_component = image[:,:,1]  
#Separating out blue component
blue_component = image[:,:,2]   

BGR = np.array([blue_component,green_component,red_component])
BGR = np.swapaxes(BGR,0,1)
BGR = np.swapaxes(BGR,1,2)

plt.subplot(2,2,1)
plt.title("red component")
plt.imshow(red_component,cmap= 'jet')
#Show the colorbar
plt.colorbar(orientation = 'horizontal')    
plt.axis('off')

plt.subplot(2,2,2)
plt.title("green component")
plt.imshow(green_component,cmap = 'jet')
#Show the colorbar
plt.colorbar(orientation = 'horizontal')    
plt.axis('off')

plt.subplot(2,2,3)
plt.title("blue component")
plt.imshow(blue_component,cmap = 'jet')
#Show the colorbar
plt.colorbar(orientation = 'horizontal')    
plt.axis('off')

plt.subplot(2,2,4)
plt.title("BGR")
plt.imshow(BGR)
#Show the colorbar
plt.colorbar(orientation = 'horizontal')    
plt.axis('off')

plt.show()
```
